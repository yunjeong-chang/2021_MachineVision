# Gaussian Filtering

```python
import cv2 as cv
from google.colab.patches import cv_imshow
img = cv.imread('/content/drive/MyDrive/MachineVision/samples/opencv-logo-white.png')
blur1 = cv.GaussianBlur(img,ksize=(7,7),sigmaX=3) 
blur2 = cv.GaussianBlur(img,ksize=(11,11),sigmaX=5)
cv_imshow(img)
cv_imshow(blur1)
cv_imshow(blur2) # 시그마를 키울 수록 블러 효과 세진다는 것 확인
```

# Median Filtering

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv_imshow

# 솔트앤페퍼 노이즈를 추가해주는 함수
def addsaltandpeppernoise(image, ratio): # ratio : 어느정도 비율로 노이즈를 추가할 것인가
  row,col,ch = image.shape
  out = np.copy(image)
  # Salt mode 
  num_salt = np.ceil(image.size * ratio)
  coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape] # 어떤 위치에 줄 것인지 (랜덤)
  out[coords] = 255 # 흰색
  # Pepper mode 
  num_pepper = np.ceil(image.size * ratio)
  coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape] # 어떤 위치에 줄 것인지 (랜덤)
  out[coords] = 0 # 검정색
  return out
```

```python
img = cv.imread('/content/drive/MyDrive/MachineVision/samples/opencv-logo-white.png')
img = addsaltandpeppernoise(img, ratio=0.01)
median_img = cv.medianBlur(img, ksize=3)
cv_imshow(img)
cv_imshow(median_img)
```
