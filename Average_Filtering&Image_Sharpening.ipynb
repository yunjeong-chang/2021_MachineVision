## Image Sharpening

```python
img = cv.imread('/content/drive/MyDrive/MachineVision/samples/lena.jpg')
cv_imshow(img)
```

```python
import numpy as np

blur = cv.blur(img, ksize=(5,5))
cv_imshow(blur)

detail = np.int32(img) - np.int32(blur)
cv_imshow(detail)

shapened_img = np.int32(img) + 5*detail
cv_imshow(shapened_img)
```
